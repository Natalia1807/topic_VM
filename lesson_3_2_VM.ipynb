###Задание (на листочке)
Почему прямые не кажутся перпендикулярными? (см.ролик)

Проверяем, являются ли прямые на самом деле перпендикулярными.
y = 3 * x + 1
y2 = (-1/3) * x + 1
условие перпедикулярности:
y = a1x + b1
y2 = a2x +b2
если a1 * a2 =-1, прямые перпендикулярны
a1 * a2 = 3 * (-1/3) = -1
Вывод: прямые перпендикулярны

```python
x = np.linspace(-5, 5, 21)
y = 3 * x + 1
y2 = (- 1 / 3) * x + 1
plt.plot(x, y)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
```

```python
пробуем ограничить масштаб. Для этого ограничиваем отображение прямых по осям x и y
```

```python
y = 3 * x + 1
y2 = (- 1 / 3) * x + 1
plt.plot(x, y)
plt.plot(x, y2)
plt.xlim(-6, 6)
plt.ylim(-4, 4)
plt.xlabel('x')
plt.ylabel('y')
```

вывод: прямые не кажутся перпендикулярными из-за разного масштаба осей координат.  

###3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.


```python
%matplotlib inline
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
r = 8
x = [i / 100 for i in range(-1000, 1000, 1)]
y1 = [np.sqrt(r**2 - i**2) for i in x]
y2 = [-i for i in y1]
plt.axis('equal')
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
a = 2
b = 3
x = [i / 100 for i in range(-1000, 1000, 1)]
y1 = [np.sqrt(b**2 * (a**2 - i**2)) / a**2 for i in x]
y2 = [-i for i in y1]
plt.axis('equal')
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
```

4. Задание (в программе)
Решите систему уравнений:
y = x2 – 1
exp(x) + x∙(1 – y) = 1


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2, 3, 201)
plt.plot(x, (np.exp(x) + x - 1) / x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,10) 
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x - x * y - 1)

x1, y1 =  fsolve(equations, (2, 4))
x2, y2 = fsolve(equations, (-2, -1))             
print(x1, y1)
print(x2, y2)
```
